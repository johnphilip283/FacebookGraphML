```python
import jgraph as jg
import random
def visualize_network(infile):
    
    network = open(infile, "r")
    
    lines = network.readlines()[:100]
    
    lines = [(word.split(' ')[0], word.split(' ')[1][:-1]) for word in lines]
    
    jg.draw(lines)

visualize_network("bfs_twitter_network_3.txt")
```
